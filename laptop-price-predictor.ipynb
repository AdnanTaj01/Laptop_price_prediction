```python
# Convert Memory to string and clean it
df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)
df['Memory'] = df['Memory'].str.replace('GB', '', regex=False)
df['Memory'] = df['Memory'].str.replace('TB', '000', regex=False)

# Split into first and second storage
new = df['Memory'].str.split('+', n=1, expand=True)
df['first'] = new[0].str.strip()
df['second'] = new[1].str.strip() if new.shape[1] > 1 else "0"

# Layer flags for first
df['Layer1HDD'] = df['first'].apply(lambda x: 1 if 'HDD' in str(x) else 0)
df['Layer1SSD'] = df['first'].apply(lambda x: 1 if 'SSD' in str(x) else 0)
df['Layer1Hybrid'] = df['first'].apply(lambda x: 1 if 'Hybrid' in str(x) else 0)
df['Layer1Flash_Storage'] = df['first'].apply(lambda x: 1 if 'Flash Storage' in str(x) else 0)

# Keep only digits in first
df['first'] = df['first'].str.replace(r'\D', '', regex=True)

# Handle second column
df['second'] = df['second'].fillna("0").astype(str)

df['Layer2HDD'] = df['second'].apply(lambda x: 1 if 'HDD' in x else 0)
df['Layer2SSD'] = df['second'].apply(lambda x: 1 if 'SSD' in x else 0)
df['Layer2Hybrid'] = df['second'].apply(lambda x: 1 if 'Hybrid' in x else 0)
df['Layer2Flash_Storage'] = df['second'].apply(lambda x: 1 if 'Flash Storage' in x else 0)

# Keep only digits in second
df['second'] = df['second'].str.replace(r'\D', '', regex=True)

# Convert to int
df['first'] = df['first'].astype(int)
df['second'] = df['second'].astype(int)

# Total storage columns
df['HDD'] = df['first']*df['Layer1HDD'] + df['second']*df['Layer2HDD']
df['SSD'] = df['first']*df['Layer1SSD'] + df['second']*df['Layer2SSD']
df['Hybrid'] = df['first']*df['Layer1Hybrid'] + df['second']*df['Layer2Hybrid']
df['Flash_Storage'] = df['first']*df['Layer1Flash_Storage'] + df['second']*df['Layer2Flash_Storage']

# Drop intermediate columns
df.drop(columns=['first','second','Layer1HDD','Layer1SSD','Layer1Hybrid',
                 'Layer1Flash_Storage','Layer2HDD','Layer2SSD','Layer2Hybrid',
                 'Layer2Flash_Storage'], inplace=True)

print("Memory preprocessing completed successfully!")
```

### Linear regression

```python
# Categorical columns in your data
cat_cols = ['Company', 'TypeName', 'Cpu brand', 'Gpu brand', 'os']

# Step 1: ColumnTransformer with OneHotEncoder
# handle_unknown='ignore' ensures new categories in test set do not break the pipeline
step1 = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), cat_cols)
    ],
    remainder='passthrough'  # numeric columns are passed as-is
)

# Step 2: Linear Regression model
step2 = LinearRegression()

# Pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Fit pipeline on training data
pipe.fit(X_train, y_train)

# Predict on test data
y_pred = pipe.predict(X_test)

# Evaluate
print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
```

### Voting Regressor

```python
from sklearn.ensemble import VotingRegressor,StackingRegressor

# Categorical columns
cat_cols = ['Company', 'TypeName', 'Cpu brand', 'Gpu brand', 'os']

# Preprocessing: OneHotEncode categorical columns
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), cat_cols)
    ],
    remainder='passthrough'
)

# Base regressors
rf = RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5,
                           max_features=0.75, max_depth=15, bootstrap=True)
gbdt = GradientBoostingRegressor(n_estimators=100, max_features=0.5)
xgb = XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5, verbosity=0)
et = ExtraTreesRegressor(n_estimators=100, random_state=3, max_samples=0.5,
                         max_features=0.75, max_depth=10, bootstrap=True)

# Voting Regressor
voting_reg = VotingRegressor(
    estimators=[('rf', rf), ('gbdt', gbdt), ('xgb', xgb), ('et', et)],
    weights=[5, 1, 1, 1]
)

# Pipeline
pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('voting', voting_reg)
])

# Fit and predict
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)

# Evaluate
print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
```

### Stacking

```python
from sklearn.ensemble import VotingRegressor,StackingRegressor

# Categorical columns
cat_cols = ['Company', 'TypeName', 'Cpu brand', 'Gpu brand', 'os']

# Preprocessing: OneHotEncode categorical columns
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), cat_cols)
    ],
    remainder='passthrough'
)

# Base regressors
rf = RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5,
                           max_features=0.75, max_depth=15, bootstrap=True)
gbdt = GradientBoostingRegressor(n_estimators=100, max_features=0.5)
xgb = XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5, verbosity=0)
et = ExtraTreesRegressor(n_estimators=100, random_state=3, max_samples=0.5,
                         max_features=0.75, max_depth=10, bootstrap=True)

# Stacking Regressor with Ridge as meta-model
stack_reg = StackingRegressor(
    estimators=[('rf', rf), ('gbdt', gbdt), ('xgb', xgb), ('et', et)],
    final_estimator=Ridge(alpha=1.0),
    passthrough=True   # allows base model predictions + original features for meta-model
)

# Pipeline
pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('stacking', stack_reg)
])

# Fit and predict
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)

# Evaluate
print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
```
